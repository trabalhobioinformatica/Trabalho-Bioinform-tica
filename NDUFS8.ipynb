```python
#gene id 4728 corresponde à base de dados "gene" na qual o gene foi revisto


from Bio import Entrez
Entrez.email="tbioinformatica@gmail.com" #always tell who you are
data=Entrez.esummary(db="gene",id=4728,retmode="xml")
info=Entrez.read(data)
ac_number=info["DocumentSummarySet"]["DocumentSummary"][0]["GenomicInfo"][0]["ChrAccVer"]
gene_start=info["DocumentSummarySet"]["DocumentSummary"][0]["GenomicInfo"][0]["ChrStart"]
gene_end=info["DocumentSummarySet"]["DocumentSummary"][0]["GenomicInfo"][0]["ChrStop"]
print ("AC Number:", ac_number,"Gene Start:",gene_start, "Gene End:", gene_end)
```

```python
#NC_00011.10 é o id do fasta que corresponde à base de dados primaria (sem anotações) 

import os
from Bio import Entrez
from Bio import SeqIO

Entrez.email="tbioinformatica@gmail.com"
handle=Entrez.efetch(db="nucleotide",rettype="gb",retmode="text",id="NC_000011.10", seq_start="68030680", seq_stop="68036643")
seq_record=SeqIO.read(handle,"gb")
SeqIO.write(seq_record,"gene.gb","genbank")
handle.close ()
record=SeqIO.read("gene.gb","genbank")
print ("REGISTO NCBI" + "\n")
print(record)

for feature in record.features:
    print(feature)
```
