```python
#NC_000002.12 é o id do fasta que corresponde à base de dados primaria (sem anotações) 

import os
from Bio import Entrez
from Bio import SeqIO

Entrez.email="tbioinformatica@gmail.com"
handle=Entrez.efetch(db="nucleotide",rettype="gb",retmode="text",id="NC_000002.12", seq_start="206159443", seq_stop="206114816")
seq_record=SeqIO.read(handle,"gb")
SeqIO.write(seq_record,"gene.gb","genbank")
handle.close ()
record=SeqIO.read("gene.gb","genbank")
print ("REGISTO NCBI" + "\n")
print(record)

for feature in record.features:
    print(feature)
```
